```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv('/Users/gmkim/Desktop/ML_Learning/Andrew_Ng_Assignments/K_means_Clustering/College_Data',index_col=0)

data.head()
```

```python
data['P.Num'] = 0
for i in range(len(data)):
    if data['Private'][i] == 'Yes':
        data['P.Num'][i] = 1
    elif data['Private'][i] == 'No':
        data['P.Num'][i] = 0
print(data)
```

```python
fig , axes = plt.subplots()
axes.scatter(x=data['Outstate'],y=data['F.Undergrad'],c=data['P.Num'])
```

```python
from sklearn.cluster import KMeans
```

```python
re_data = data.drop('Private',axis=1)
kmeans = KMeans(n_clusters=2)
kmeans.fit(re_data)

# 실루엣 스코어 : optimized n_cluster 도출
```

```python
fig , axes = plt.subplots()
axes.scatter(x=re_data['Outstate'],y=re_data['F.Undergrad'],c=kmeans.labels_)
```

```python
fig, (axes1, axes2) = plt.subplots(nrows=1,ncols=2)
axes1.scatter(x=data['Outstate'],y=data['F.Undergrad'],c=data['P.Num'])
axes1.set_title("Original")
axes1.set_xlabel("Outstate")
axes1.set_ylabel("F.Undergrad")
axes2.scatter(x=re_data['Outstate'],y=re_data['F.Undergrad'],c=kmeans.labels_)
axes2.set_title("K-means")
axes2.set_xlabel("Outstate")
axes2.set_ylabel("F.Undergrad")
fig.tight_layout()
```
