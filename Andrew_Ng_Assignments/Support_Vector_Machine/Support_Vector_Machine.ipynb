```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from mat4py import loadmat
```

```python
# Read the data from .mat file
data = loadmat('/Users/gmkim/Desktop/ML_Learning/Andrew_Ng_Assignments/Support_Vector_Machine/ex6data2.mat')
df = pd.DataFrame(data['X'][:],columns=['x1','x2'])
flat_y = list(np.concatenate(data['y']).flat)
df['y'] = flat_y
```

```python
print(df)
print(df.shape)
print(len(df.index))
```

```python
X = np.array(df[['x1','x2']])
y = np.array(df['y'])
print(X)
print(y)
```

```python
import sklearn.svm as svm
```

```python
def plot_svc_decision_function(model, ax=None, plot_support=True):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 30)
    y = np.linspace(ylim[0], ylim[1], 30)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='k',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--'])
    
    # plot support vectors
    if plot_support:
        ax.scatter(model.support_vectors_[:, 0],
                   model.support_vectors_[:, 1],
                   s=300, linewidth=1, facecolors='none');
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)
```

```python
sv = svm.SVC(kernel='rbf', C=200)
result = sv.fit(X,y)
print(result)
```

```python
for i in range(len(df.index)):
    if y[i] == 1:
        plt.scatter(X[i,0],X[i,1],color='blue',marker='o')
    elif y[i] == 0:
        plt.scatter(X[i,0],X[i,1],color='red',marker='^')

plot_svc_decision_function(result)
plt.xlabel('x1')
plt.ylabel('x2')

plt.show()
```
